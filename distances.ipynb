```python
googleMapsURL = "https://routes.googleapis.com/directions/v2:computeRoutes?key=AIzaSyD3Rlnx2e02ab6L7YBvRrvVDU-rBPZ_Las"
headers = {"X-Goog-FieldMask": "*"}

college_paths = {college: {} for college in collegeList.keys()}

for college1, address1 in collegeList.items():
    for college2, address2 in collegeList.items():
        requestBody = {
            "origin": {
                "address": address1
            },
            "destination": {
                "address": address2
            },
            "travelMode": "WALK"
        }
        r = requests.post(googleMapsURL, headers=headers, json=requestBody)

        college_paths[college1][college2] = r.json()
```
